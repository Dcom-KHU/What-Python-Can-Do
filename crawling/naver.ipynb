# 네이버 실시간 검색어 랭킹 크롤링
현재 네이버의 실시간 검색어 랭킹을 크롤링 한다.

```python
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get("https://www.naver.com")
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
arr = list(map(lambda x: x.text, soup.select('span.ah_k')))
```

```python
print("현재 검색어 순위")
for i in range(20):
    print(i+1, "위 :", arr[i])
```
